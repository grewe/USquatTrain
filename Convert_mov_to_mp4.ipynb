```
# Define paths 
MP4_OUTPUT_DIR = '/content/drive/MyDrive/Dataset/MP4_OUTPUT'
SUB_FOLDERS = ['good','knees_in','shallow','knees_in','knees_forward','heels_off']

for item in SUB_FOLDERS:
  # Create path for each sub folders
  sub_path = os.path.join(DATASET_DIR,item)
  mp4_sub_path = os.path.join(MP4_OUTPUT_DIR,item)
  # if mp4_output does not exists, create a folder
  if not os.path.isdir(mp4_sub_path):
    print('directory not Exists')
    os.mkdir(mp4_sub_path)
  else:
    print('Directory already exists')
    for video in os.listdir(sub_path):
      input_path = os.path.join(sub_path,video)
      if '.mov' in input_path.lower():
        output_path = os.path.join(mp4_sub_path,video.lower().replace('.mov','.mp4'))
        # convert .mov to .mp4 using stream copy 
        # https://ffmpeg.org/ffmpeg.html#Stream-copy
        os.system(f"ffmpeg -i {input_path} -c copy {output_path}")
      else:
        # can move .mp4 to new folder using Google Drive API or do it manually
        print('files already in mp4')
```
